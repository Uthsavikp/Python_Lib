```python
''' 
  @Author: Uthsavi KP
  @Date: 2021-01-15 21:58:27
  @Last Modified by: Uthsavi KP
  @Last Modified time: 2021-01-15 21:58:27
  @Title: To draw a scatter plot of “day” against “total bill” for a dataset.
'''
import pandas as pd
import seaborn as sns

class ScatterPlot:
    @staticmethod
    def scatter_plot_of_day_vs_total_bill():
        """
        drawing a scatterplot of day against total bill
        """
        data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
        sns.scatterplot(x = "day", y = "total_bill",data = data)

if __name__=="__main__":
    ScatterPlot.scatter_plot_of_day_vs_total_bill()
```
