```python
''' 
  @Author: Uthsavi KP
  @Date: 2021-01-15 21:58:27
  @Last Modified by: Uthsavi KP
  @Last Modified time: 2021-01-15 21:58:27
  @Title: To draw a violin plot of “species” against “sepal length” for iris dataset
'''
import pandas as pd
import seaborn as sns

class ViolinPlot:
    @staticmethod
    def violin_plot_for_species_against_sepal_length():
        """
        drawing a violin plot for species against sepal length
        """
        data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
        sns.violinplot(x = "species", y = "sepal_length", data = data)

if __name__=="__main__":
    ViolinPlot.violin_plot_for_species_against_sepal_length()
```
