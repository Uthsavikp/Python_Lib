```python
''' 
  @Author: Uthsavi KP
  @Date: 2021-01-15 23:48:24
  @Last Modified by: Uthsavi KP
  @Last Modified time: 2021-01-15 23:48:24
  @Title: To draw a swarm plot of total bill against day for tips dataset.
'''
import pandas as pd
import seaborn as sns

class SwarmPlot:
    @staticmethod
    def total_bill_against_day():
        """
        drawing a swarm plot of total bills against day
        """
        data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
        sns.swarmplot(x = "total_bill", y = "day", data = data)
if __name__=="__main__":
    SwarmPlot.total_bill_against_day()
```
