```python
''' 
  @Author: Uthsavi KP
  @Date: 2021-01-15 22:33:02
  @Last Modified by: Uthsavi KP
  @Last Modified time: 2021-01-15 22:33:02
  @Title: To draw box plot of life expectancy by continent for a dataset.
'''
import pandas as pd
import seaborn as sns

class BoxPlot:
    @staticmethod
    def box_plot_for_life_expectancy():
        """
        drawing a box plot for lixe expectancy by continent wise
        """
        data = pd.read_csv("https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv")
        sns.boxplot(x = "lifeExp", y = "continent", palette=["b","y","r","w","c"], data = data)
if __name__=="__main__":
    BoxPlot.box_plot_for_life_expectancy()
```
