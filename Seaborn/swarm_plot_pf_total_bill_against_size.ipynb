```python
''' 
  @Author: Uthsavi KP
  @Date: 2021-01-15 23:36:16
  @Last Modified by: Uthsavi KP
  @Last Modified time: 2021-01-15 23:36:16
  @Title: To draw a swarm plot of total bill against size for tips dataset.
'''
import pandas as pd
import seaborn as sns

class SwarmPlot:
    @staticmethod
    def total_bill_against_size():
        """
        drawing a swarm plot of total bills against size
        """
        data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
        sns.swarmplot(x = "size", y = "total_bill", data = data)
if __name__=="__main__":
    SwarmPlot.total_bill_against_size()
```
