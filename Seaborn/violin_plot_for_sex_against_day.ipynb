```python
''' 
  @Author: Uthsavi KP
  @Date: 2021-01-15 21:58:27
  @Last Modified by: Uthsavi KP
  @Last Modified time: 2021-01-15 21:58:27
  @Title: To draw a violin plot of sex against day for tips dataset.
'''
import pandas as pd
import seaborn as sns

class ViolinPlot:
    @staticmethod
    def violin_plot_of_sex_against_day():
        """
        drawing a voilin plot for sex against day,
        since both sex and day columns are categorical,
        using one in hue.
        """
        data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
        sns.violinplot(x = "day", y = "tip", hue = "sex", data = data)

if __name__=="__main__":
    ViolinPlot.violin_plot_of_sex_against_day()
```
