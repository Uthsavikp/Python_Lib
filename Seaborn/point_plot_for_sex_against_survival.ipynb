```python
''' 
  @Author: Uthsavi KP
  @Date: 2021-01-15 21:21:33
  @Last Modified by: Uthsavi KP
  @Last Modified time: 2021-01-15 21:40:12
  @Title: To draw a point plot for sex against survived for titanic dataset.
'''
import pandas as pd
import seaborn as sns

class PointPlot:
    @staticmethod
    def point_plot_showing_sex_against_survived():
        """
        drawing a pointplot for sex againt 
        survival in titanic dataset
        """
        data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv")
        sns.pointplot(x = "sex", y = "survived",data = data)

if __name__=="__main__":
    PointPlot.point_plot_showing_sex_against_survived()
```
